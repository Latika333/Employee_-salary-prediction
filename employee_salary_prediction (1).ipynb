```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler, OneHotEncoder

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

models = {
    "LogisticRegression": LogisticRegression(),
    "RandomForest": RandomForestClassifier(),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(),
    "GradientBoosting": GradientBoostingClassifier()
}

results = {}

for name, model in models.items():
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('model', model)
    ])

    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc
    print(f"{name} Accuracy: {acc:.4f}")
    print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
plt.bar(results.keys(), results.values(), color='skyblue')
plt.ylabel('Accuracy Score')
plt.title('Model Comparison')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import joblib

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Define models
models = {
    "LogisticRegression": LogisticRegression(max_iter=1000),
    "RandomForest": RandomForestClassifier(),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(),
    "GradientBoosting": GradientBoostingClassifier()
}

results = {}

# Train and evaluate
for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds)
    results[name] = acc
    print(f"{name}: {acc:.4f}")

# Get best model
best_model_name = max(results, key=results.get)
best_model = models[best_model_name]
print(f"\n✅ Best model: {best_model_name} with accuracy {results[best_model_name]:.4f}")

# Save the best model
joblib.dump(best_model, "best_model.pkl")
print("✅ Saved best model as best_model.pkl")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Sample synthetic dataset
data = {
    'Occupation': np.random.choice(['Tech', 'Sales', 'Admin', 'Manager'], 200),
    'Age': np.random.randint(20, 60, 200),
    'Income': np.random.normal(70000, 25000, 200)
}

df = pd.DataFrame(data)

# Create Age groups
bins = [18, 29, 45, 100]
labels_age = ['<30', '30-45', '>45']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels_age, right=True)

# Create Income brackets
bins_income = [0, 50000, 100000, np.inf]
labels_income = ['<50k', '50k-100k', '>100k']
df['IncomeGroup'] = pd.cut(df['Income'], bins=bins_income, labels=labels_income, right=False)

# Pivot table: counts of people by Occupation and IncomeGroup for each AgeGroup
grouped = df.groupby(['Occupation', 'AgeGroup', 'IncomeGroup']).size().unstack(fill_value=0)

# Colors for income groups
income_colors = {'<50k': '#e41a1c', '50k-100k': '#377eb8', '>100k': '#4daf4a'}

# Define hatch patterns for visual differentiation
hatch_patterns = {'<50k': '', '50k-100k': '//', '>100k': 'xx'}

fig, ax = plt.subplots(figsize=(12, 7))
bar_height = 0.25
occupations = ['Tech', 'Sales', 'Admin', 'Manager']

num_age_groups = len(labels_age)
total_bar_height = bar_height * num_age_groups
y_pos = np.arange(len(occupations)) * (total_bar_height + 0.3)

for i, age_group in enumerate(labels_age):
    if age_group not in grouped.index.get_level_values('AgeGroup'):
        continue
    sub_data = grouped.loc[(slice(None), age_group), :].reset_index(level=1, drop=True)
    sub_data = sub_data.reindex(occupations, fill_value=0)
    left = np.zeros(len(occupations))

    for income_group in labels_income:
        if income_group in sub_data.columns:
            counts = sub_data[income_group].values
            bars = ax.barh(
                y_pos + i * bar_height,
                counts,
                bar_height,
                left=left,
                color=income_colors[income_group],
                edgecolor='black',
                linewidth=0.7,
                alpha=0.85,
                hatch=hatch_patterns[income_group],
                label=f"{income_group}" if i == 0 else None,
            )
            left += counts

ax.set_yticks(y_pos + total_bar_height / 2 - bar_height / 2)
ax.set_yticklabels(occupations, fontsize=12)

ax.set_xlabel('Count of People', fontsize=14)
ax.set_title('Occupation vs Income Group Stacked by Age Group', fontsize=16)
ax.legend(title='Income Groups', fontsize=12, title_fontsize=13)

plt.tight_layout()
plt.show()
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import joblib
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

# --- 1. Load the trained model and create necessary preprocessors ---
# The model was saved as 'best_model.pkl'
model = joblib.load("best_model.pkl")

# To correctly preprocess new data, we need the LabelEncoders and StandardScaler
# that were fitted on the training data.
# Since they weren't explicitly saved, we'll re-create them by fitting on a dummy
# dataset that mimics the original data's unique values and structure.
# In a real project, you would save these preprocessor objects directly.

# Dummy data to fit encoders and scaler (mimicking original data's unique values)
# This is crucial for consistent preprocessing.
# Based on the notebook's data cleaning and encoding steps:

# Create a dummy DataFrame with all possible unique values for categorical columns
# and representative values for numerical columns.
# This is a simplified representation; ideally, you'd save the fitted encoders/scalers.
dummy_data = pd.DataFrame({
    'age': [30, 17, 75], # Min, max, and a typical age
    'workclass': ['Private', 'Local-gov', 'Others', 'Self-emp-not-inc', 'State-gov', 'Self-emp-inc', 'Federal-gov'],
    'fnlwgt': [100000, 200000, 300000], # Representative values
    'educational-num': [7, 9, 12, 10, 6, 15, 5, 16], # All educational-num values after filtering
    'marital-status': ['Never-married', 'Married-civ-spouse', 'Widowed', 'Divorced', 'Separated', 'Married-AF-spouse', 'Married-spouse-absent'],
    'occupation': ['Machine-op-inspct', 'Farming-fishing', 'Protective-serv', 'Others', 'Other-service', 'Prof-specialty', 'Craft-repair', 'Exec-managerial', 'Adm-clerical', 'Sales', 'Transport-moving', 'Handlers-cleaners', 'Tech-support', 'Priv-house-serv', 'Armed-Forces'],
    'relationship': ['Own-child', 'Husband', 'Not-in-family', 'Unmarried', 'Wife', 'Other-relative'],
    'race': ['Black', 'White', 'Asian-Pac-Islander', 'Amer-Indian-Eskimo', 'Other'],
    'gender': ['Male', 'Female'],
    'capital-gain': [0, 7688, 15024], # Representative values
    'capital-loss': [0, 1000, 2000], # Representative values
    'hours-per-week': [40, 50, 30, 38, 10, 99, 1], # Representative values
    'native-country': ['United-States', 'Mexico', 'Philippines', 'Germany', 'Canada', 'England', 'Puerto-Rico', 'El-Salvador', 'India', 'Cuba', 'Jamaica', 'South', 'China', 'Italy', 'Dominican-Republic', 'Vietnam', 'Guatemala', 'Columbia', 'Poland', 'Japan', 'Portugal', 'Haiti', 'Taiwan', 'Iran', 'Greece', 'Nicaragua', 'Peru', 'Ecuador', 'France', 'Ireland', 'Thailand', 'Hong', 'Cambodia', 'Trinadad&Tobago', 'Laos', 'Yugoslavia', 'Outlying-US(Guam-USVI-etc)', 'Honduras', 'Hungary', 'Scotland', 'Holand-Netherlands']
})

# Initialize LabelEncoders for each categorical column
label_encoders = {}
categorical_cols_for_encoding = ['workclass', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'native-country']
for col in categorical_cols_for_encoding:
    le = LabelEncoder()
    # Fit on unique values from the dummy data to ensure all possible categories are known
    le.fit(dummy_data[col].astype(str).unique()) # Convert to string to handle potential NA if dummy_data had them
    label_encoders[col] = le

# Initialize StandardScaler
scaler = StandardScaler()
# Fit scaler on numerical columns from the dummy data
numerical_cols = ['age', 'fnlwgt', 'educational-num', 'capital-gain', 'capital-loss', 'hours-per-week']
scaler.fit(dummy_data[numerical_cols])


# --- 2. Streamlit App Layout ---
st.set_page_config(page_title="Employee Salary Classification", page_icon="💼", layout="centered")

st.title("💼 Employee Salary Classification App")
st.markdown("Predict whether an employee earns >50K or ≤50K based on input features.")

# --- 3. Sidebar for User Input ---
st.sidebar.header("Input Employee Details")

# Numerical Inputs
age = st.sidebar.slider("Age", 17, 75, 30) # Adjusted min/max based on outlier removal
fnlwgt = st.sidebar.number_input("Final Weight (fnlwgt)", min_value=10000, max_value=1500000, value=200000)
educational_num = st.sidebar.slider("Educational Number (Years of Education)", 5, 16, 9) # Adjusted min/max based on outlier removal
capital_gain = st.sidebar.number_input("Capital Gain", min_value=0, max_value=100000, value=0)
capital_loss = st.sidebar.number_input("Capital Loss", min_value=0, max_value=5000, value=0)
hours_per_week = st.sidebar.slider("Hours per week", 1, 99, 40) # Adjusted min/max based on outlier removal

# Categorical Inputs (using selectbox for predefined options)
workclass = st.sidebar.selectbox("Workclass", dummy_data['workclass'].unique())
marital_status = st.sidebar.selectbox("Marital Status", dummy_data['marital-status'].unique())
occupation = st.sidebar.selectbox("Occupation", dummy_data['occupation'].unique())
relationship = st.sidebar.selectbox("Relationship", dummy_data['relationship'].unique())
race = st.sidebar.selectbox("Race", dummy_data['race'].unique())
gender = st.sidebar.selectbox("Gender", dummy_data['gender'].unique())
native_country = st.sidebar.selectbox("Native Country", dummy_data['native-country'].unique())


# --- 4. Prepare Input for Prediction ---
# Create a DataFrame from the current user inputs
input_data = pd.DataFrame({
    'age': [age],
    'workclass': [workclass],
    'fnlwgt': [fnlwgt],
    'educational-num': [educational_num],
    'marital-status': [marital_status],
    'occupation': [occupation],
    'relationship': [relationship],
    'race': [race],
    'gender': [gender],
    'capital-gain': [capital_gain],
    'capital-loss': [capital_loss],
    'hours-per-week': [hours_per_week],
    'native-country': [native_country]
})

st.write("### 🔎 Input Data")
st.write(input_data)

# --- 5. Preprocess Input Data (matching notebook's steps) ---
processed_input_data = input_data.copy()

# Apply Label Encoding
for col in categorical_cols_for_encoding:
    # Use .transform() on the fitted encoder. Handle unknown categories if they appear.
    # For simplicity, we assume user selects from known categories.
    processed_input_data[col] = label_encoders[col].transform(processed_input_data[col])

# Apply StandardScaler
processed_input_data[numerical_cols] = scaler.transform(processed_input_data[numerical_cols])


# --- 6. Prediction ---
if st.button("Predict Salary Class"):
    try:
        prediction = model.predict(processed_input_data)
        # The model predicts either '<=50K' or '>50K' directly
        st.success(f"✅ Predicted Salary Class: **{prediction[0]}**")
    except Exception as e:
        st.error(f"An error occurred during prediction: {e}")
        st.write("Please ensure all inputs are valid and match the expected format.")

# --- 7. Batch Prediction ---
st.markdown("---")
st.markdown("#### 📂 Batch Prediction")
uploaded_file = st.file_uploader("Upload a CSV file for batch prediction", type="csv")

if uploaded_file is not None:
    batch_data = pd.read_csv(uploaded_file)
    st.write("Uploaded data preview:", batch_data.head())

    # Ensure batch data has the same columns as training data (excluding 'income')
    # and preprocess it similarly.
    try:
        # Drop 'education' if present in batch data
        if 'education' in batch_data.columns:
            batch_data = batch_data.drop(columns=['education'])

        # Apply Label Encoding to batch data
        for col in categorical_cols_for_encoding:
            # Handle potential unknown categories in batch data
            # For simplicity, we'll replace unknown with the mode of the original training data
            # A more robust solution would involve a custom transformer or error handling
            if col in batch_data.columns:
                # Check for values in batch_data[col] that are not in label_encoders[col].classes_
                unknown_mask = ~batch_data[col].isin(label_encoders[col].classes_)
                if unknown_mask.any():
                    st.warning(f"Found unknown categories in '{col}' in batch file. Replacing with mode.")
                    # Find the mode from the original encoder's classes
                    mode_val = label_encoders[col].inverse_transform([label_encoders[col].transform([label_encoders[col].classes_[0]])[0]])[0] # Get the first class as a fallback mode
                    batch_data.loc[unknown_mask, col] = mode_val

                batch_data[col] = label_encoders[col].transform(batch_data[col])
            else:
                st.warning(f"Column '{col}' not found in uploaded batch file. Skipping encoding for this column.")

        # Apply StandardScaler to batch data
        if all(col in batch_data.columns for col in numerical_cols):
            batch_data[numerical_cols] = scaler.transform(batch_data[numerical_cols])
        else:
            st.error("Missing numerical columns in batch file for scaling.")
            st.stop() # Stop execution if critical columns are missing

        batch_preds = model.predict(batch_data)
        batch_data['PredictedClass'] = batch_preds
        st.write("✅ Predictions:")
        st.write(batch_data.head())
        csv = batch_data.to_csv(index=False).encode('utf-8')
        st.download_button("Download Predictions CSV", csv, file_name='predicted_classes.csv', mime='text/csv')

    except Exception as e:
        st.error(f"An error occurred during batch prediction: {e}")
        st.write("Please ensure your CSV file has the correct columns and data types.")

```
